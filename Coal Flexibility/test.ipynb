```python
df = pd.read_csv("../data/BW01.csv",sep='\t')
df.index.name = 'Date'
df.index = pd.to_datetime(df.index)
df.sort_index(inplace=True)

df = df.diff() / 5

up = df[df>0]

down = df[df<0]

ax = up.plot(alpha=0.9)

down.plot(ax=ax,alpha=0.9)
```

```python
df = pd.read_csv("../data/BW01.csv",sep='\t')
df.index.name = 'Date'
df.index = pd.to_datetime(df.index)

new = df.copy()
ramp = df.copy()

new.columns = ['seasonal']
ramp.columns = ['ramp rate']
df.columns = ['daily']


df = df.resample('1h').sum().pct_change()
df = df[df<20] # make this less than the maximum of seasonal
ax = df.plot(figsize=(10,6),grid=True)

ramp.index = df.index
ramp = ramp.diff()/5
ramp.plot(ax=ax)

new = new.resample('ME').mean().pct_change()
new = new[new<20]
new.plot(ax=ax)
plt.show()
```
